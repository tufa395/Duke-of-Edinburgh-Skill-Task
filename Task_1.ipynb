```python
import pandas as pd
from IPython.display import display

# Step 1: Read the CSV
df = pd.read_csv("students.csv")

print("âœ… Step 1: CSV Loaded Successfully")
print("Total Rows in Dataset:", len(df))
print("\nFirst 5 rows:")
display(df.head())
```

```python
# Step 2: Validation function
def validate_row(row):
    reasons = []
    # Rule 1: score <= max_score
    if row["score"] > row["max_score"]:
        reasons.append("score greater than max_score")

    # Rule 2: score >= 0 and max_score > 0
    if row["score"] < 0:
        reasons.append("negative score")
    if row["max_score"] <= 0:
        reasons.append("max_score must be > 0")
    # Rule 3: valid date
    try:
        pd.to_datetime(row["date"], format="%Y-%m-%d", errors="raise")
    except:
        reasons.append("invalid date format")
    return "; ".join(reasons)
# Apply validation
df["reason"] = df.apply(validate_row, axis=1)
print("âœ… Step 2: Validation Applied")
print("\nValidation Reasons Column Added:")
display(df[["student_id","name","assessment_type","score","max_score","date","reason"]])
```

```python
# Step 3: Separate invalid & valid rows
invalid_df = df[df["reason"] != ""]
valid_df   = df[df["reason"] == ""].drop(columns=["reason"])

# Save invalid rows to CSV
invalid_df.to_csv("invalid_rows.csv", index=False)

print("âœ… Step 3: Invalid Rows Separated and Saved as invalid_rows.csv")
print("Valid Rows Count:", len(valid_df))
print("Invalid Rows Count:", len(invalid_df))

print("\nðŸ”µ VALID DATA")
display(valid_df)

print("\nðŸ”´ INVALID DATA")
display(invalid_df)
```

```python
#PART-B STEP 1: Calculate Percentage
# Step 1: Percentage
valid_df["percentage"] = (valid_df["score"] / valid_df["max_score"]) * 100

print("âœ… Step 1: Percentage Calculated")
display(valid_df[["student_id","name","assessment_type","score","max_score","percentage"]])
```

```python
# Step 2: Extract ISO week number
valid_df["week"] = pd.to_datetime(valid_df["date"]).dt.isocalendar().week

print("âœ… Step 2: ISO Week Number Added")
display(valid_df[["student_id","name","date","week"]])
```

```python
# Step 3: Assessment Weight
weights_map = {
    "Quiz": 0.2,
    "Assignment": 0.3,
    "Mid": 0.2,
    "Final": 0.3
}

def get_weight(atype):
    if atype in weights_map:
        return weights_map[atype]
    else:
        print(f"âš ï¸ Warning: Unknown assessment_type '{atype}' found. Weight set to 0.")
        return 0

valid_df["assessment_weight"] = valid_df["assessment_type"].apply(get_weight)

print("âœ… Step 3: Assessment Weights Assigned")
display(valid_df[["student_id","name","assessment_type","percentage","week","assessment_weight"]])
```

```python
#Part C â€” Course-wise Final Grade Calculation

import numpy as np

# Step 1: Compute weighted score per row
valid_df["weighted_score"] = valid_df["percentage"] * valid_df["assessment_weight"]

# Step 2: Aggregate by student_id + course
final_df = valid_df.groupby(["student_id", "name", "course"], as_index=False)["weighted_score"].sum()
final_df = final_df.rename(columns={"weighted_score":"final_score"})

# Step 3: Assign Letter Grades
def assign_grade(score):
    if score >= 90: return "A+"
    elif score >= 85: return "A"
    elif score >= 80: return "A-"
    elif score >= 75: return "B+"
    elif score >= 70: return "B"
    elif score >= 60: return "C"
    else: return "F"

final_df["grade"] = final_df["final_score"].apply(assign_grade)

# Step 4: Save final results
final_df.to_csv("final_results.csv", index=False)

print("âœ… Part-C Completed: Final course grades calculated and saved to final_results.csv")
display(final_df)
```

```python
#PART-D: Analytics Report
# Step 1: Open a file for writing
with open("report.txt", "w") as f:

    f.write("ðŸ“Š STUDENT PERFORMANCE ANALYTICS REPORT\n")
    f.write("="*50 + "\n\n")

    courses = final_df["course"].unique()

    for course in courses:
        f.write(f"Course: {course}\n")
        f.write("-"*40 + "\n")

        course_df = final_df[final_df["course"] == course]

        # 1. Top 3 students
        f.write("Top 3 Students:\n")
        top3 = course_df.sort_values(by="final_score", ascending=False).head(3)
        for i, row in top3.iterrows():
            f.write(f"{row['name']} (ID: {row['student_id']}): {row['final_score']:.2f}\n")
        f.write("\n")

        # 2. Course statistics
        mean_score = course_df["final_score"].mean()
        median_score = course_df["final_score"].median()
        std_score = course_df["final_score"].std()
        f.write(f"Course Statistics:\nMean: {mean_score:.2f}, Median: {median_score:.2f}, Std Dev: {std_score:.2f}\n\n")

        # 3. Grade distribution
        f.write("Grade Distribution:\n")
        grade_counts = course_df["grade"].value_counts()
        for grade, count in grade_counts.items():
            f.write(f"{grade}: {count}\n")
        f.write("\n")

        # 4. At-risk students
        f.write("At-risk Students (F or < 60):\n")
        at_risk = course_df[(course_df["final_score"] < 60) | (course_df["grade"] == "F")]
        if at_risk.empty:
            f.write("None\n")
        else:
            for i, row in at_risk.iterrows():
                f.write(f"{row['name']} (ID: {row['student_id']}): {row['final_score']:.2f}, Grade: {row['grade']}\n")
        f.write("\n" + "="*50 + "\n\n")

print("âœ… Part-D Completed: report.txt generated successfully!")

# Display first few lines of the report
with open("report.txt", "r") as f:
    lines = f.readlines()
    print("".join(lines[:20]))  # first 20 lines
```

```python
#PART-E: StudentAnalyzer Class + CLI Tool

# Colab-compatible StudentAnalyzer

import pandas as pd
import os

class StudentAnalyzer:
    def __init__(self, input_file, outdir):
        self.input_file = input_file
        self.outdir = outdir
        os.makedirs(outdir, exist_ok=True)
        self.df = None
        self.valid_df = None
        self.invalid_df = None
        self.final_df = None

    def load_and_clean(self):
        self.df = pd.read_csv(self.input_file)
        def validate_row(row):
            reasons = []
            if row["score"] > row["max_score"]:
                reasons.append("score greater than max_score")
            if row["score"] < 0:
                reasons.append("negative score")
            if row["max_score"] <= 0:
                reasons.append("max_score must be > 0")
            try:
                pd.to_datetime(row["date"], format="%Y-%m-%d", errors="raise")
            except:
                reasons.append("invalid date format")
            return "; ".join(reasons)

        self.df["reason"] = self.df.apply(validate_row, axis=1)
        self.invalid_df = self.df[self.df["reason"] != ""]
        self.valid_df = self.df[self.df["reason"] == ""].drop(columns=["reason"])
        self.invalid_df.to_csv(os.path.join(self.outdir,"invalid_rows.csv"), index=False)
        print(f"âœ… Invalid rows saved: {len(self.invalid_df)}")

    def compute_results(self):
        weights_map = {"Quiz":0.2, "Assignment":0.3, "Mid":0.2, "Final":0.3}
        self.valid_df["percentage"] = (self.valid_df["score"] / self.valid_df["max_score"])*100
        self.valid_df["week"] = pd.to_datetime(self.valid_df["date"]).dt.isocalendar().week

        def get_weight(atype):
            if atype in weights_map:
                return weights_map[atype]
            else:
                print(f"âš ï¸ Unknown assessment_type '{atype}' found. Weight set to 0.")
                return 0

        self.valid_df["assessment_weight"] = self.valid_df["assessment_type"].apply(get_weight)
        self.valid_df["weighted_score"] = self.valid_df["percentage"] * self.valid_df["assessment_weight"]

        self.final_df = self.valid_df.groupby(["student_id","name","course"], as_index=False)["weighted_score"].sum()
        self.final_df = self.final_df.rename(columns={"weighted_score":"final_score"})

        def assign_grade(score):
            if score >= 90: return "A+"
            elif score >= 85: return "A"
            elif score >= 80: return "A-"
            elif score >= 75: return "B+"
            elif score >= 70: return "B"
            elif score >= 60: return "C"
            else: return "F"

        self.final_df["grade"] = self.final_df["final_score"].apply(assign_grade)
        self.final_df.to_csv(os.path.join(self.outdir,"final_results.csv"), index=False)
        print(f"âœ… Final results saved: {len(self.final_df)}")

    def generate_report(self):
        with open(os.path.join(self.outdir,"report.txt"),"w") as f:
            f.write("ðŸ“Š STUDENT PERFORMANCE REPORT\n")
            f.write("="*50+"\n\n")

            courses = self.final_df["course"].unique()
            for course in courses:
                f.write(f"Course: {course}\n")
                f.write("-"*40 + "\n")
                course_df = self.final_df[self.final_df["course"]==course]

                f.write("Top 3 Students:\n")
                top3 = course_df.sort_values(by="final_score", ascending=False).head(3)
                for i,row in top3.iterrows():
                    f.write(f"{row['name']} (ID:{row['student_id']}): {row['final_score']:.2f}\n")
                f.write("\n")

                mean_score = course_df["final_score"].mean()
                median_score = course_df["final_score"].median()
                std_score = course_df["final_score"].std()
                f.write(f"Course Stats: Mean={mean_score:.2f}, Median={median_score:.2f}, Std={std_score:.2f}\n\n")

                f.write("Grade Distribution:\n")
                grade_counts = course_df["grade"].value_counts()
                for grade,count in grade_counts.items():
                    f.write(f"{grade}: {count}\n")
                f.write("\n")

                f.write("At-risk Students (F or <60):\n")
                at_risk = course_df[(course_df["final_score"]<60)|(course_df["grade"]=="F")]
                if at_risk.empty:
                    f.write("None\n")
                else:
                    for i,row in at_risk.iterrows():
                        f.write(f"{row['name']} (ID:{row['student_id']}): {row['final_score']:.2f}, Grade:{row['grade']}\n")
                f.write("\n" + "="*50 + "\n\n")
        print("âœ… Report generated: report.txt")

# -----------------------------
# Colab Execution
# -----------------------------
input_file = "students.csv"  # directly specify input
output_dir = "output"        # output directory

analyzer = StudentAnalyzer(input_file, output_dir)
analyzer.load_and_clean()
analyzer.compute_results()
analyzer.generate_report()
```
